Correlation explains how one or more variables are related to each other. These variables can be input data features which have been used to forecast our target variable. Correlation, statistical technique which determines how one variables moves/changes in relation with the other variable.

HEAT MAP :
A heat map represents these coefficients to visualize the strength of correlation among variables. It helps find features that are best for Machine Learning model building. The heat map transforms the correlation matrix into color coding.

```python
corelation = movies.corr()
sns.heatmap(corelation, xticklabels = corelation.columns, yticklabels = corelation.columns)
```

Pairplot visualizes given data to find the relationship between them where the variables can be continuous or categorical.

```python
sns.pairplot(movies)
```

```python
the hue parameter represents which column in the data frame, you want to use for color encoding.
```

```python
sns.relplot(x = "movie_id", y = "popularity",hue = "budget", data = movies)
```
